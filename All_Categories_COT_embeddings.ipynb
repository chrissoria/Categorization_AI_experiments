Defining the function so that I can loop through the UCNets data

```python
def create_embedding(client, text, model="text-embedding-3-large", encoding_format="float"):
    response = client.embeddings.create(
        model=model,
        input=text,
        encoding_format=encoding_format
    )
    return response

def get_embedding_for_row(text):
    try:
        embedding_response = create_embedding(client, text)
        return embedding_response.data[0].embedding
    except Exception as e:
        print("Error in getting embedding:", e)
        return None
```

```python
embeddings = UCNets['a19i'].apply(get_embedding_for_row)
```
